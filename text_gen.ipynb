```python
# STEP 2: Import Libraries
import torch
from transformers import GPT2Tokenizer, GPT2LMHeadModel
```

```python
# STEP 3: Load GPT-2 Model and Tokenizer
MODEL_ID = "gpt2"
tokenizer_gpt = GPT2Tokenizer.from_pretrained(MODEL_ID)
model_gpt = GPT2LMHeadModel.from_pretrained(MODEL_ID)
model_gpt.eval()
```

```python
# STEP 4: Define Text Generation Function
def create_text(seed_text, max_tokens=150, temp=0.7, nucleus=0.9):
    encoded_input = tokenizer_gpt.encode(seed_text, return_tensors='pt')
    with torch.no_grad():
        generated_output = model_gpt.generate(
            encoded_input,
            max_length=max_tokens,
            temperature=temp,
            top_p=nucleus,
            do_sample=True,
            num_return_sequences=1,
            pad_token_id=tokenizer_gpt.eos_token_id
        )
    final_output = tokenizer_gpt.decode(generated_output[0], skip_special_tokens=True)
    return final_output
```

```python
# STEP 5: Run the Generator
if __name__ == "__main__":
    print("🚀 GPT-2 Prompt Generator")
    user_input = input("Type a prompt or topic: ")
    result = create_text(user_input)
    print("\n✨ Generated Content:\n")
    print(result)
```
